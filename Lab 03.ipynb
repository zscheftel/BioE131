```python
readIn = open("romeoandjuliet.txt")
```

```python
uniqueWords = {}
for word in readIn.read().split():
    if word not in uniqueWords:
        uniqueWords[word] = 1
    else:
        uniqueWords[word] += 1
for k, v in uniqueWords.items():
    print(k, v)
```

```python
import matplotlib.pylab as plt
import operator
import numpy as np
import math
```

```python
sortedDict = sorted(uniqueWords.items(), key=operator.itemgetter(1))
```

```python
print(sortedDict)
```

```python
x, y = zip(*sortedDict)
```

```python
print(x, y)
```

```python
plt.bar(range(len(x)), y, 0.75)
plt.xlabel("Words", fontsize = 20)
plt.xticks(range(len(x)), x)
plt.ylabel("Frequency", fontsize = 20)
plt.ylim(0, 400)
plt.rcParams['figure.figsize'] = [15, 15]
plt.show()
```

The ten most frequent words are "the", "I", "and", "to", "a", "of", "my", "is", "in", and "that". This suggests that the text is most often using articles, pronouns, and prepositions.
The ten least frequent words are all words that occur just once in the text. These words are quite varied, ranging from proper nouns, to adjectives, to nouns.

```python
invertedX = x[::-1]
invertedY = y[::-1]
fig1, ax1 = plt.subplots()

plt.plot(range(len(x)), invertedY, 0.75)
plt.xlabel("Words", fontsize = 20)
ax1.set_xscale('log')
ax1.set_yscale('log')
plt.ylabel("Frequency", fontsize = 20)
plt.rcParams['figure.figsize'] = [15, 15]
plt.show()
```

This graph suggests that the frequency of the use of words in the text decreases logarithmically. This is an example of Zipf's Law.
